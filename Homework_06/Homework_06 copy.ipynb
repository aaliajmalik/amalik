Use the `.groupby()` method to group the data by species and calculate the average and standard deviation for each variable based on the iris species.

```python
grouped_data = df_new.groupby('Species')
average_data = grouped_data.mean()
std_data = grouped_data.mean()

print("Average data by species:")
print(average_data)

print("\nStandard deviation data by species:")
print(std_data)
```
