```python
def partial_cor(cor,x,y,z):
    from math import sqrt
    rdk = ((cor[x][y]-cor[x][z]*cor[y][z])/ 
       (sqrt(1-cor[x][z]**2)*sqrt(1-cor[y][z]**2)))
    return rdk
```

```python
def partial_cor_modified(cor_matrix, variable):
    l = []  # Initialize empty list l

    # Iterate through each variable a in cor_matrix
    for a in cor_matrix.columns[2:]:
        # Skip if a is the same as the control variable
        if a == variable:
            continue
        # Iterate through each variable b in cor_matrix
        for b in cor_matrix.columns[2:]:
            # Skip if b is the same as the control variable or same as a
            if b == variable or b == a:
                continue
            # Calculate partial correlation and append to list l
            l.append(partial_cor(cor_matrix, a, b, variable))

    return l
```
